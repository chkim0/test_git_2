```python
# a. 숫자 하나를 입력받는다
# b. 1부터 입력받은 숫자까지 범위에서 짝수만 선택
# c. (b)의 숫자 각각 팩토리얼 (!n) 연산 결과를 구해 출력한다

#출력예시  
# 6을 입력한 경우 
# 2 , 24, 720 

# 팩토리얼 (!)
# * 팩토리얼의 정의 숫자 n이있을때 n*(n-1)*(n-2)...*1
```

```python
x = int(input())
```

```python
# 
res = 1

for i in range(1,x+1):
    #모든 i를 곱해서 저장해야 한다
    res *= i
    print(res)
```

```python
for i in range(2,x+1,2):
    
    res = 1

    for j in range(1,i+1):
    #모든 i를 곱해서 저장해야 한다
        res *= j    # j * (j+1) * (j+2) * (j+3)...
    print(res)
```
