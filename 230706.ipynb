```python
# 함수 실습
# 함수이름은 func1
# 입력받는값은 (x,y)
# 반환값 x / y 의 나머지와 x - y 를 비교해 
# 전자가 크면 Type 1 
# 둘이같거나 후자가 크면 Type ii라고 반환
```

```python
def func1(x, y):
    if x % y > x - y :
        return 'Type 1'
    elif x % y <= x - y:                                         # 연습 
        return 'Type 2'
```

```python
func1(22,3)
```
